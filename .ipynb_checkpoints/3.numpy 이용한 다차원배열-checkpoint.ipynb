- 맷플롯립을 사용한 일반 도표
```python
plt.plot(y)
plt.plot(x,y)
```

````python
x=np.linspace(-5.,5.,100)
y=np.sin(3*x)*np.exp(-.1*x**2)
plt.plot(x,y,'--^',lw=3,color='#fdbb84',mfc='#2b8cbe',ms=8) 
# --는 점선스타일 ,^ 삼각형 명시
# lw는 선의굵기
# mfc는 마커 색상
# ms 는 마커 사이즈
````

````python
x=np.random.randn(100)
y=x+np.random.randn(100)
plt.scatter(x,y) 
````

# 맷플롯립 도표 수정

````python
plt.subplot(1,2,1)
x= np.linspace(-10.,10.,1000)
plt.plot(x,np.sin(x),'-r',label='sinus')
plt.plot(x,np.cos(x),':g',lw=1,label='cosinus')
plt.xticks([-10,0,10])
plt.yticks([-1,0,1])
plt.ylim(-2,2)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Two plots")
plt.legend()
````

````python
from ipywidgets import interact
x=np.linspace(-5.,5.,1000)
@interact
def plot_sin(a=(1,10)):
    plt.plot(x,np.sin(a*x))
    plt.ylim(-1,1)
````
